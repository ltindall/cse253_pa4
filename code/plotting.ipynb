```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure()
loss = np.load('LSTM_1lay_100unit_25seq_1000batch_200epoch_0drop_1temp.npy')
loss2 = np.load('LSTM_1lay_50unit_25seq_1000batch_200epoch_0drop_1temp.npy')
loss3 = np.load('LSTM_1lay_150unit_25seq_1000batch_200epoch_0drop_1temp.npy')
plt.plot(loss[0,:],label='Train 100 units')
plt.plot(loss2[0,:], label='Train 50 units')
plt.plot(loss3[0,:], label='Train 150 units')
plt.plot(loss[1,:],label='Validation 100 units')
plt.plot(loss2[1,:], label = 'Validation 50 units')
plt.plot(loss3[1,:], label = 'Validation 150 units')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title("Loss vs Epochs")
plt.legend()
plt.show()
```
