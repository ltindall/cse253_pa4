```python
import numpy as np
import matplotlib.pyplot as plt
## Different Units
loss = np.load('LSTM_1lay_100unit_25seq_1000batch_500epoch_0drop_1temp.npy')
plt.figure()
plt.plot(loss[0,:],'--',label='Train')
plt.plot(loss[1,:],label='Validation')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title("Loss vs Epochs, 100 units, 1 layer, 0 dropout, Temp=1")
plt.legend()
plt.show()
```

```python
## Different Units
loss = np.load('LSTM_1lay_100unit_25seq_1000batch_500epoch_0drop_1temp.npy')
loss1=np.load('LSTM_1lay_75unit_25seq_1000batch_500epoch_0drop_1temp.npy')
loss2 = np.load('LSTM_1lay_50unit_25seq_1000batch_500epoch_0drop_1temp.npy')
loss3 = np.load('LSTM_1lay_150unit_25seq_1000batch_500epoch_0drop_1temp.npy')
plt.figure()
plt.plot(loss[0,:],'--',label='Train 100 units')
plt.plot(loss1[0,:],'--',label='Train 75 units')
plt.plot(loss2[0,:],'--', label='Train 50 units')
plt.plot(loss3[0,:],'--', label='Train 150 units')
plt.plot(loss[1,:],label='Validation 100 units')
plt.plot(loss1[1,:],label='Validation 75 units')
plt.plot(loss2[1,:], label = 'Validation 50 units')
plt.plot(loss3[1,:], label = 'Validation 150 units')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title("Loss vs Epochs")
plt.legend()
plt.show()
```

```python
# Different dropout
loss = np.load('LSTM_1lay_100unit_25seq_1000batch_500epoch_0drop_1temp.npy')
loss1=np.load('LSTM_1lay_100unit_25seq_1000batch_500epoch_0.1drop_1temp.npy')
loss2 = np.load('LSTM_1lay_100unit_25seq_1000batch_500epoch_0.2drop_1temp.npy')
loss3 = np.load('LSTM_1lay_100unit_25seq_1000batch_500epoch_0.3drop_1temp.npy')
plt.figure()
plt.plot(loss[0,:],'--',label='Train 0 dropout')
plt.plot(loss1[0,:],'--',label='Train 0.1 dropout')
plt.plot(loss2[0,:],'--', label='Train 0.2 dropout')
plt.plot(loss3[0,:],'--', label='Train 0.3 dropout')
plt.plot(loss[1,:],label='Validation 0 dropout')
plt.plot(loss1[1,:],label='Validation 0.1 dropout')
plt.plot(loss2[1,:], label = 'Validation 0.2 dropout')
plt.plot(loss3[1,:], label = 'Validation 0.3 dropout')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title("Loss vs Epochs")
plt.legend()
plt.show()
```
